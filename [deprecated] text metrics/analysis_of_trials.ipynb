```
import pandas as pd

# Load trials.csv into a pandas DataFrame
df_trials = pd.read_csv('/content/drive/MyDrive/Clubs/BEA/LLM Project/dataverse_files/trials.csv')

# Load schema.txt into a string
with open('/content/drive/MyDrive/Clubs/BEA/LLM Project/dataverse_files/schema.txt', 'r') as f:
    schema_content = f.read()

# Print the first few rows of the DataFrame
print("First 5 rows of df_trials:")
print(df_trials.head())

# Print the content of schema.txt
print("\nContent of schema.txt:")
print(schema_content)
```

**Reasoning**:

```
print("\n--- DataFrame Info ---")
df_trials.info()

print("\n--- Missing Values (Percentage) ---")
missing_values = df_trials.isnull().sum()
missing_percentage = (missing_values / len(df_trials)) * 100
missing_df = pd.DataFrame({
    'Missing Count': missing_values,
    'Missing Percentage': missing_percentage
})
print(missing_df[missing_df['Missing Count'] > 0].sort_values(by='Missing Percentage', ascending=False))

# Extract expected column names from schema.txt
schema_lines = schema_content.split('\n')
expected_columns = []
for line in schema_lines:
    if ',' in line and not line.startswith(('version', '@totalColumns', 'Variable name')):
        col_name = line.split(',')[0].strip()
        expected_columns.append(col_name)

print("\n--- Column Name Consistency ---")
current_columns = set(df_trials.columns)
expected_columns_set = set(expected_columns)

missing_in_df = expected_columns_set - current_columns
extra_in_df = current_columns - expected_columns_set

if missing_in_df:
    print(f"Columns present in schema but not in DataFrame: {missing_in_df}")
else:
    print("All expected columns from schema are present in the DataFrame.")

if extra_in_df:
    print(f"Columns present in DataFrame but not in schema: {extra_in_df}")
else:
    print("No extra columns found in DataFrame compared to schema.")
```
